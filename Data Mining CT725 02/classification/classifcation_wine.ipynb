```python
X = df.drop(columns=["quality"])
y = df["quality"]
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
```

```python
X_train.shape, X_test.shape
```

```python
y_train.shape, y_test.shape
```

### Model 1: Linear Regression

```python
from sklearn.linear_model import LinearRegression

lr_model = LinearRegression()

lr_model.fit(X_train, y_train)
```

```python
y_preds_lr = lr_model.predict(X_test)
y_preds_lr[:10]
```

```python
from sklearn.metrics import mean_squared_error, r2_score

# Evaluate predictions
mse = mean_squared_error(y_test, y_preds_lr)
r2 = r2_score(y_test, y_preds_lr)

print("Mean Squared Error:", mse)
print("R-squared:", r2)
```

### Model 2: Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

log_model = LogisticRegression()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
log_model.fit(X_train_scaled, y_train)
```

```python
y_preds_log = log_model.predict(X_test_scaled)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

# Assuming y_test contains the true labels and y_preds_log contains the predicted labels
accuracy = accuracy_score(y_test, y_preds_log)

print("Accuracy:", accuracy)
```

## Model 3: Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt_model = DecisionTreeClassifier()
dt_model.fit(X_train, y_train)

y_preds_dt = dt_model.predict(X_test)
```

```python
from sklearn.metrics import classification_report

print(accuracy_score(y_test, y_preds_dt))
print(f1_score(y_test, y_preds_dt,average='weighted'))
print(classification_report(y_test, y_preds_dt))
```

### Model 4: Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier()
rf_model.fit(X_train, y_train)

y_preds_rf = rf_model.predict(X_test)
```

```python
print(accuracy_score(y_test, y_preds_rf))
print(f1_score(y_test, y_preds_rf,average='weighted'))
print(classification_report(y_test, y_preds_rf))
```

### Model 5: KNN

```python
from sklearn.neighbors import KNeighborsClassifier

knn_model = KNeighborsClassifier(n_neighbors=7)
knn_model.fit(X_train, y_train)

y_preds_knn = knn_model.predict(X_test)

print(accuracy_score(y_test, y_preds_knn))
print(f1_score(y_test, y_preds_knn,average='weighted'))
print(classification_report(y_test, y_preds_knn))
```

```python
n_estimators = [20, 60, 100, 120]

max_features = [0.2, 0.6, 1.0]

max_depth = [2, 8, None]

max_samples = [0.5, 0.75, 1.0]
```

```python
param_grid = {
    'n_estimators': n_estimators,
    'max_features': max_features,
    'max_depth': max_depth,
    'max_samples': max_samples
}
```

```python
from sklearn.model_selection import GridSearchCV

rf_grid = GridSearchCV(
    estimator = RandomForestClassifier(),
    param_grid = param_grid,
    cv = 5,
    verbose = 2,
    n_jobs = -1
)
```

```python
rf_grid.fit(X_train, y_train)
```

```python
rf_grid.best_params_
```

```python
rf_best = RandomForestClassifier(n_estimators=100, max_features=0.2, max_depth=None, max_samples=1.0)

rf_best.fit(X_train, y_train)

y_preds_rf_best = rf_best.predict(X_test)

accuracy_rf_best = accuracy_score(y_test, y_preds_rf_best)

accuracy_rf_best
```
