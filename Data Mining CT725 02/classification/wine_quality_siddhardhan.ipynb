```python
X = wine_dataset.drop('quality', axis=1)
y = wine_dataset['quality']
```

```python
X
```

```python
y.head()
```

```python
Y = wine_dataset["quality"].apply(lambda y_value: 1 if y_value >= 7 else 0)
```

```python
Y.value_counts()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=3)
```

```python
X_train.shape, X_test.shape
```

```python
def  split_data(test_size, X, y):
    test_start_idx = int(test_size * len(X))
    test = X[:test_start_idx], y[:test_start_idx]
    train = X[test_start_idx:], y[test_start_idx:]
    return train, test
```

```python
train, test = split_data(0.2, X, Y)
```

```python
train[0].shape, train[1].shape, test[0].shape, test[1].shape
```

```python
model = RandomForestClassifier()

model.fit(X_train, y_train)
```

```python
y_preds_rf = model.predict(X_test)
```

```python
y_preds_rf[:20]
```

```python
accuracy = accuracy_score(y_test, y_preds_rf)
```

```python
accuracy
```

```python
from sklearn.model_selection import cross_val_score
np.mean(cross_val_score(model, X, Y, cv=5, scoring='accuracy'))
```

### Grid search CV

```python
n_estimators = [20, 60, 100, 120]

max_features = [0.2, 0.6, 1.0]

max_depth = [2, 8, None]

max_samples = [0.5, 0.75, 1.0]
```

```python
param_grid = {
    'n_estimators': n_estimators,
    'max_features': max_features,
    'max_depth': max_depth,
    'max_samples': max_samples
}
```

```python
print(param_grid)
```

```python
from sklearn.model_selection import GridSearchCV

rf_grid = GridSearchCV(
    estimator = RandomForestClassifier(),
    param_grid = param_grid,
    cv = 5,
    verbose = 2,
    n_jobs = -1
)
```

```python
rf_grid.fit(X_train, y_train)
```

```python
rf_grid.best_params_
```

```python
rf_best = RandomForestClassifier(n_estimators=100, max_features=0.2, max_depth=None, max_samples=1.0)

rf_best.fit(X_train, y_train)

y_preds_rf_best = rf_best.predict(X_test)

accuracy_rf_best = accuracy_score(y_test, y_preds_rf_best)

accuracy_rf_best
```
